## Plotting Using Seaborn and PyPlot

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

### Pairplot

```python
sns.pairplot(porsche[["Price", "Age", "Mileage"]])
plt.show()
```

### Vertical Barchart

```python
plt.figure();
porsche.plot(kind = 'bar', stacked = True);
plt.show()
```

### Horizontal Barchart

```python
porsche.plot(kind='barh', stacked=True);
plt.show()
```

### Histogram

```python
plt.figure();
porsche['Mileage'].diff().hist(bins = 7)
plt.show()
```

### Andrews Curves

```python
from pandas.tools.plotting import andrews_curves
plt.figure()
andrews_curves(porsche, 'Age', colormap='winter')
plt.show()
```

```python
plt.figure()
andrews_curves(porsche, 'Age', colormap='winter')
plt.show()
```

### Radial Visualization

```python
from pandas.tools.plotting import radviz

plt.figure()
radviz(porsche, 'Age')
plt.show()
```
